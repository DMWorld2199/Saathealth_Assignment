```python
# Importing all the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
 
# Opening JSON file
data = open('SampleData(15Nov2018_to_13Sep2019.json')
 
# returns JSON object as
# a dictionary
data_dict = json.load(data)

print(data_dict['-LRHbArGBwe-KuIpvVqq']['user_id'])
```

```python
# Converting the JSON format into dataframe format 
df = pd.DataFrame(data_dict)
df = df.transpose()  
df
```

```python
# Replacing the categorical index of dataframe with numerical index
df.reset_index(inplace = True, drop = True)
df
```

```python
# Changing the epoch timestamp to Date format
df['paid_at'] = pd.to_datetime(df['paid_at'], unit='ms')
df['redeemed_at'] = pd.to_datetime(df['redeemed_at'], unit='ms')
df
```

# Task 1:  Converting all timestamp of each redemption to a proper Date format (yyyy-mm-dd)

```python
df['paid_at'] = df['paid_at'].dt.date
df['redeemed_at'] = df['redeemed_at'].dt.date
df
```

```python
# Removing day from timestamp and adding new column with only Month and Year

df['Time'] = df['redeemed_at'].apply(lambda x: x.strftime('%Y-%m'))
df
```

```python
times = list(set((df["Time"].values)))
times
```

# Task 2: Code for creating No. of transactions per User for period between Nov 2018 to Sep 2019

```python
from collections import defaultdict
userInfo = defaultdict(lambda: defaultdict(int))
userInfo[1780]["2011 - 18"]
```

```python
for i in df.itertuples():
  time = i.Time
  userId = i.user_id
  userInfo[userId][time]+=1

userIds = list(set(df["user_id"].values))

rows = []
for user in userIds:
  cur = []
  for t in times:
    cur.append(userInfo[user][t])
  rows.append(cur)

customerData = pd.DataFrame(rows, columns=times)

customerData["user_id"] = userIds
```

```python
customerData.set_index("user_id", inplace = True)
customerData
```

```python
customerData.to_csv('customerData.csv')
```

## a. Transactions that are done by each merchant for the 10 months period in total

```python
merchants = list(set(df["k_user_id"].values))
merchantToUser = defaultdict(lambda: defaultdict(list))

sum = 0
for i in df.itertuples():
  merchantToUser[i.k_user_id][i.Time].append(i.paid_amount)

rows = []
for m in merchants:
  cur = []
  for t in times:
    cur.append(merchantToUser[m][t])
  rows.append(cur)

merchanttrans = pd.DataFrame(rows, columns=times)
merchanttrans["merchant_id"] = merchants
merchanttrans.set_index("merchant_id", inplace = True)
```
